```python
# RSI (Relative Strength Index)
def calculate_rsi(data, window):
    delta = data['close'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()
    rs = gain / loss
    rsi = 100 - (100 / (1 + rs))
    return rsi
```

```python
for i,ticker in enumerate(tickers):
    start_date = globals()[f"df{i}"].iloc[-121,0]
    end_date = globals()[f"df{i}"].iloc[-1,0]
    exdf = globals()[f"df{i}"].iloc[-121:-1]
    # 일일 수익률
    exdf['daily_return'] = exdf['close'].pct_change()
    exdf['moving_avg_20'] = exdf['close'].rolling(window=20).mean()
    exdf['moving_avg_50'] = exdf['close'].rolling(window=50).mean()
    # 볼린저 밴드
    exdf['stddev'] = exdf['close'].rolling(window=20).std()
    exdf['upper_band'] = exdf['moving_avg_20'] + (exdf['stddev'] * 2)
    exdf['lower_band'] = exdf['moving_avg_20'] - (exdf['stddev'] * 2)
    # ATR (Average True Range)
    exdf['high_low'] = exdf['high'] - exdf['low']
    exdf['high_close'] = np.abs(exdf['high'] - exdf['close'].shift())
    exdf['low_close'] = np.abs(exdf['low'] - exdf['close'].shift())
    exdf['tr'] = exdf[['high_low', 'high_close', 'low_close']].max(axis=1)
    exdf['atr_14'] = exdf['tr'].rolling(window=14).mean()
    exdf['rsi_14'] = calculate_rsi(exdf, 14)
    # 불필요한 컬럼 제거
    exdf = exdf.drop(columns=['moving_avg_20','moving_avg_50','high_low', 'high_close', 'low_close', 'tr', 'stddev'])
    display(exdf)
```
